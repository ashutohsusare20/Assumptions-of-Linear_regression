# Assumptions

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Assumption are main topic in LR, by validating this assumption we can make sure that our algorithm model is Good or Not

>> If all the assumption is validate by the model then the Goodness of model is very high.
>> If some assumption get viloadted then model is not good to predict accurate value.

>> For model improvement, you also need to understand regression assumptions and ways to fix them when they get violated.
```

```python
There are five basic assumption of linear regression :-
    
    1) Linearity:- (Linear relationship between x and mean y)
    2) No multiCo-linearity:- (independent columns should not be linear with each other)
    3) Normality of residual:- (residual error should be normally distributed)
            1) kdeplot
            2) QQ plot
            3) skewness of residual
            4) Hypothesis testing:-
                        1) Shapiro
                        2) ksTest
                        3) normaltest
        
    4) Homoscedasticity :- (residual error should be constant deviation from the zero line)
    5) No Autocorrelation of error:- (sctter plot of error should not be in fix pattern)
```

## 2) No multicollinearity :-

```python
The independent variables should not be correlated.

>> If there is co-rrelation between the independent variabls then this called as multicollinearity.

>> We can check multicollinearity by using VIF.

>> VIF ( Variance Inflation Factor ):- It is used to check the multicollinarity in regression analysis.
```

```python
VIF equal to 1 = variables are not correlated
VIF between 1 and 5 = variables are moderately correlated 
VIF greater than 5 = variables are highly correlated
```

### 4) Hypothesis Testing :- 

```python
Data is normally distributed -----> Null Hypothesis
Data is not normally distributed ----->Alternate Hypothesis
```

```python
p_val Range ----> 0 to 1

(0.05) is significant value in Hypothesis testing


p_val >= 0.05    >> We are accepting NULL hypothesis   (more or equal to 0.05)
p_Val < 0.05     >> We are accepting Alternate Hypothesis (less than 0.05)

p_val >= 0.05    >> NULL hypothesis is True
p_Val < 0.05     >> NULL hypothesis is False / Alternate Hypothesis is True
```
